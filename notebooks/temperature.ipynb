# Temperature

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/temperature.csv")
df
```

```python
plt.figure(figsize=(14,8))
plt.plot(df.ds, df.y, color="royalblue", label="Values")
plt.plot(df.ds, df.y.rolling(50).mean(), color="black", linestyle=":", label="Rolling mean - 50")
plt.plot(df.ds, df.y.rolling(100).mean(), color="black", linestyle="--", label="Rolling mean - 100")
plt.xlabel("Points")
plt.ylabel("Value")
plt.legend()
plt.show()

plt.figure(figsize=(14,8))
plt.plot(df.ds, df.y.rolling(50).std(), color="olivedrab", linestyle=":", label="Rolling std - 50")
plt.plot(df.ds, df.y.rolling(100).std(), color="olivedrab", linestyle="--", label="Rolling std - 100")
plt.xlabel("Points")
plt.ylabel("Value")
plt.legend()
plt.show()
```
